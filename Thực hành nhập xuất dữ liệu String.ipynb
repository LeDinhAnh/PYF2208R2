##bài 4
nhập vào 1 'string' và một số 'x'. hãy xóa 'x' ký tự đầu tiên của string
ví dụ : string = " Hello Python", x = 3
output: "lo python"

```python
string = input(" nhập vào một chuỗi")
x = int(input(" nhập x"))
string = string[x:]
print(f"String sau khi cắt là: {string}")

```
